```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import online data
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
# input variables ya features (X)
digits.data.shape
X = digits.data
# means 1797 pictures size 64=8x8
```

```python
# output labels (y)
digits.target.shape
y = digits.target
```

```python
plt.figure(figsize=(20,4))
for index, (image, label) in enumerate(zip(digits.data[0:5], digits.target[0:5])):
    plt.subplot(1, 5, index + 1)
    plt.imshow(np.reshape(image, (8,8)), cmap=plt.cm.gray)
    plt.title("Training: %i\n" % label, fontsize = 20)
```

```python
plt.figure(figsize=(20,4))
for index, (image, label) in enumerate(zip(digits.data[0:20], digits.target[0:20])):
    plt.subplot(1, 20, index + 1)
    plt.imshow(np.reshape(image, (8,8)), cmap=plt.cm.gray)
    plt.title(label, fontsize = 20)
```

```python
# help(plt)
```

```python
# split the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state=0)
```

```python
X_train.shape
```

```python
print("Train input data: ",X_train.shape)
print("Test input data: ",X_test.shape)
print("Train output data: ",y_train.shape)
print("Test output data: ",y_test.shape)
```

```python
# model train
from sklearn.linear_model import LogisticRegression

model = LogisticRegression().fit(X_train, y_train)
model
```

```python
model.predict(X_test[0:10])
predictions = model.predict(X_test)
predictions
```

```python
# acuracy test
score = model.score(X_test, y_test)
print("The accuracy score is:", score)
```

```python
# confusion matrix
from sklearn import metrics

cm = metrics.confusion_matrix(y_test, predictions)
cm
```

```python
import seaborn as sns
plt.figure(figsize=(9,9))
sns.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = "Spectral");
plt.ylabel("Actual Output");
plt.xlabel("Predicted Output");
all_sample_title = "Accuracy Score: {0}".format(score)
plt.title(all_sample_title, size = 15);
```

```python
print(cm)
```

```python
# Getting miss classified labels
import numpy as np
import matplotlib.pyplot as plt
index = 0
misclassifiedIndexes = []
for label, predict in zip(y_test, predictions):
    if label != predict:
        misclassifiedIndexes.append(index)
        index +=1
```

```python
# plotting missclassified labels with know labels
plt.figure(figsize=(20,4))
for plotIndex, badIndex in enumerate(misclassifiedIndexes[5:10]):
    plt.subplot(1, 5, plotIndex + 1)
    plt.imshow(np.reshape(X_test[badIndex], (8,8)), cmap=plt.cm.gray)
    plt.title("Predicted: {}, Actual: {}".format(predictions[badIndex], y_test[badIndex]), fontsize = 15)
```
