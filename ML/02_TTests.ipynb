# Student's T-tests
1. One-sample t-test
2. Two sample t-test
   1. Un-paird or Independent t-test
   2. Paired or relational/dependent t-test

**One-sample Student's t-test**

Test a sample with a know standard value.

**Assumptions**

* Observations in sample is independent and indentically distributed.
* Observations in sample is normally distributed.

**Interpretation**

> **H0:** the mean of sample is equal to the known value.\
> **H1:** the mean of sample is unequal to the known value.

```python
# One sample t-test

import seaborn as sns
import pandas as pd
from scipy.stats import ttest_1samp

# load dataset

df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df1 = df[['sex', 'age', 'fare']]
df1.head()
```

```python
# Data description
df1.describe()
```

```python
# Check the age and compare with a known value of 45 years

ttest_1samp(df1['fare'], 50)
stat, p = ttest_1samp(df1['fare'], 50)

print('stat=%.3f, p=%.3f' % (stat, p))

# Adding a conditional argument for ease
if p > 0.05:
    print('Probably the same Distributions')
else:
    print('Probably different Distributions')
```

## Two-sample T-tests

**Independent student's t-test**

**Assumptions**

* Observations in each sample are independent and indentically distributed.
* Observations in each sample are normally distributed.
* Observations in each sample have the same varience.

**Interpretation**

> **H0:** the means of the samples are equal.\
> **H1:** the means of the samples are unequal.

```python
# We will compare age and fare of male Vs female passengers
# Splitting datasets

df_male = df1.loc[df1['sex']=='male']
df_female = df1.loc[df1['sex']=='female']

# library
from scipy.stats import ttest_ind
stat, p = ttest_ind(df_male['fare'], df_female['fare'])

print('stat=%.3f, p=%.3f' % (stat, p))

# Adding a conditional argument for ease
if p > 0.05:
    print('Probably the same Distributions')
else:
    print('Probably different Distributions')
```

## Paired Student's T-test

Tests whether the means of two paired samples are significantly different.
**Assumptions**

* Observations in each sample are independent and indentically distributed.
* Observations in each sample are normally distributed.
* Observations in each sample have the same varience.
* Observations across each sample are paired.

**Interpretation**

> **H0:** the means of the samples are equal.\
> **H1:** the means of the samples are unequal.

```python
df.head()
```

```python
# Select only male's data
df_m = df.loc[df['sex']=='male']
df_m.head()
```

```python
# select only two 
df_male_first = df_m.loc[df_m['class']=='First']
df_male_second = df_m.loc[df_m['class']=='Second']
df_male_third = df_m.loc[df_m['class']=='Third']
```

```python
# Check our data
df_male_third.describe()
```

```python
df_1st = df_male_first.sample(n=100)
df_2nd = df_male_second.sample(n=100)
df_3rd = df_male_third.sample(n=100)

print('The number of instances in 1st class are =', df_1st.shape)
print('The number of instances in 2nd class are =', df_2nd.shape)
print('The number of instances in 3rd class are =', df_3rd.shape)
```

```python
# Import library 

from scipy.stats import ttest_rel

# Apply test to compare class-1 and class-2
stat, p = ttest_rel(df_1st['age'],df_2nd['age'])
print('stat=%.3f, p=%.3f' % (stat, p))

# Adding a conditional argument for ease
if p > 0.05:
    print('Probably the same Distributions')
else:
    print('Probably different Distributions')
```
